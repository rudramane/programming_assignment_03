```python
# 3.Write a Python Program to Check Leap Year?
```

```python
c = int(input("Enter year : "))

if ((c % 4) or (c % 400)) == 0 :
    print(f"{c} is a leap year")
else:
    print(f"{c} is not a leap year")
```
